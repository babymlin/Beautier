```python
import pandas as pd
from glob import glob
data_list= glob("data_json/*.json")
data_list
```

```python
import json
data = []
for line in data_list:
    with open(line, "r", encoding="utf-8-sig") as file:
        data.append(json.load(file))
```

```python
for i in range(10):
    print(data[i])
    print("="*30)
```

```python
with open("form1.json", "r", encoding="utf-8-sig") as file:
    form1 = json.load(file)
form1
```

```python
print(form1["contents"][0]["hero"]["url"] + "\n" + data[2]["pic_url"])
form1["contents"][0]["hero"]["url"] = data[2]["pic_url"]
```

```python
print(form1["contents"][0]["body"]["contents"][0]["text"] + "\n" + data[2]["product_name_score"])
form1["contents"][0]["body"]["contents"][0]["text"] = data[0]["product_name_score"]
```

```python
print(form1["contents"][0]["body"]["contents"][2]["contents"][0]["contents"][0]["text"] + "\n" + data[0]["comment"])
form1["contents"][0]["body"]["contents"][2]["contents"][0]["contents"][0]["text"] = data[0]["comment"]
```

```python
print(form1["contents"][0]["footer"]["contents"][0]["action"]["uri"] + "\n" + data[0]["product_url"])
form1["contents"][0]["footer"]["contents"][0]["action"]["uri"] = data[0]["product_url"]
```
